```python
aple = pd.read_csv(company_list[0])
amzn = pd.read_csv(company_list[1])
gogl = pd.read_csv(company_list[2])
micsft = pd.read_csv(company_list[3])
```

```python
closing_price = pd.DataFrame()
```

```python
closing_price['apple_close'] = aple['close']
closing_price['amazon_close'] = amzn['close']
closing_price['google_close'] = gogl['close']
closing_price['microsoft_close'] = micsft['close']
```

```python
closing_price
```

```python
sns.pairplot(closing_price)
```

```python
closing_price.corr()
```

```python
sns.heatmap(closing_price.corr(),annot = True)
```

```python
closing_price
```

```python
closing_price['apple_close'].shift(1)
```

```python
(closing_price['apple_close'] - closing_price['apple_close'].shift(1))/closing_price['apple_close'].shift(1) * 100
```

```python
for col in closing_price.columns:
    closing_price[col + '_pct_change'] = (closing_price[col] - closing_price[col].shift(1))/closing_price[col].shift(1) * 100
```

```python
closing_price
```

```python
clsing_p_change = closing_price[['apple_close_pct_change', 'amazon_close_pct_change',
       'google_close_pct_change', 'microsoft_close_pct_change']]
```

```python
clsing_p_change
```

```python
g = sns.PairGrid(data= clsing_p_change)
g.map_diag(sns.histplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot)
```
